```python
# If not already installed, you can install them using pip:
# !pip install pandas numpy matplotlib seaborn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: make plots look better
sns.set_style('whitegrid')
%matplotlib inline
```

```python
# Replace 'your_file.csv' with the actual file path
df = pd.read_csv('idea_submissions.csv')

# Quick view of the data
df.head()
```

```python
df.columns.tolist()
```

```python
# List the columns you want to delete
columns_to_delete = ['id', 'name', 'email', 'phone','document_url',
 'supporting_url',
 'created_at',
 'objective',
 'stakeholders',
 'implementation',
 'financial_structure',
 'economic_outcomes']

# Drop them from the DataFrame
df.drop(columns=columns_to_delete, inplace=True)

# Check the result
df.head()
```

```python
# Shape of the dataset
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

# Data types and non-null counts
df.info()

# Statistical summary of numeric columns
df.describe()

# Check for missing values
df.isnull().sum()

# Check for duplicated rows
df.duplicated().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap="YlGnBu")
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Zone distribution
df['zone'].value_counts().plot(kind='bar', figsize=(10,5), color='skyblue')
plt.title('Projects per Zone')
plt.xlabel('Zone')
plt.ylabel('Number of Projects')
plt.xticks(rotation=45)
plt.show()

# State distribution
df['author_state'].value_counts().plot(kind='bar', figsize=(10,5), color='skyblue')
plt.title('Projects per State')
plt.xlabel('Zone')
plt.ylabel('Number of Projects')
plt.xticks(rotation=45)
plt.show()


# Sectors distribution
df['sectors'].value_counts().plot(kind='bar', figsize=(10,5), color='lightgreen')
plt.title('Projects per Sector')
plt.xlabel('Sector')
plt.ylabel('Number of Projects')
plt.xticks(rotation=45)
plt.show()

# Install wordcloud if needed: !pip install wordcloud
from wordcloud import WordCloud

text = " ".join(desc for desc in df['description'].dropna())

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('WordCloud for Descriptions')
plt.show()
```
